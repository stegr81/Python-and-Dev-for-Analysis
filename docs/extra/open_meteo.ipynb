```python
import requests
```

```python
url = (
    "https://api.open-meteo.com/v1/forecast"
    "?latitude=51.51&longitude=-0.13"
    "&current=temperature_2m,wind_speed_10m"
)

response = requests.get(url)
data = response.json()

current = data["current"]
print("Temperature:", current["temperature_2m"], "°C")
print("Wind speed: ", current["wind_speed_10m"], "km/h")
```

```python
# limited hourly forecast, by default the API returns 7 days of hourly data.

params = {
	"latitude": 52.52,
	"longitude": 13.41,
	"hourly": ["temperature_2m", "wind_speed_10m"],
}
```

```python
forecast = requests.get(url,params=params).json()
```

```python
# data exploration shows the construct of the returned JSON is a list of dictionaries, with the first dictionary containing the hourly data.

the_data = forecast[0]
the_hourly_data = the_data['hourly']
```

```python
# shows us the keys of our hourly data dictionary, which are the parameters we requested in our API call.

the_hourly_data.keys()
```

```python
# choosing one of the keys gives us the data, a list of values for that parameter, in this case the temperature at 2 meters above ground level.

the_hourly_data['temperature_2m']
```

```python
# we can use the past_days parameter to get historical data, in this case the last 7 days of hourly data.

historic_params = {
	"latitude": 52.52,
	"longitude": 13.41,
	"hourly": ["temperature_2m", "wind_speed_10m"],
    "past_days": 7,
}
```

```python
historic = requests.get(url, params=historic_params).json()
```
